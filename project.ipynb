```python
import pandas as pd
```

```python
raw_data = pd.read_csv("science.csv", sep='\t')
```

```python
# remove non english tweets
idx_to_delete = raw_data[raw_data["language"] != "en"].index
raw_data.drop(idx_to_delete, inplace=True)
```

```python
raw_data.filter(["hashtags", "tweet"])
```

```python
#Remove hashtags from text
import re
def remove_tag_from_text(text):
    return re.sub(r'#\w+', " ", text)
raw_data["tweet"] = raw_data["tweet"].apply(lambda row: remove_tag_from_text(row))
```

```python
raw_data.to_csv("science_cleaned.csv")
```
